```python
import pandas as pd
import numpy as np
```

```python
#Membaca file dengan sheetname Data Aktifitas dan header pada baris ke-4
df = df = pd.read_excel ('data-set.xlsx', sheet_name='Data Aktifitas', header=3)
```

```python
df.head()
```

```python
#Mengisi data NaN dengan data diatasnya
df.ffill(inplace=True)
```

```python
df.head()
```

```python
#Merubah format kolom Tanggal agar mudah diubah menjadi format datetime
df['Tanggal'] = df['Tanggal'].str.replace(' Januari ','01')
df['Tanggal'] = df['Tanggal'].str.replace(' Februari ','02')
df['Tanggal'] = df['Tanggal'].str.replace(' Maret ','03')
df['Tanggal'] = df['Tanggal'].str.replace(' April ','04')
df['Tanggal'] = df['Tanggal'].str.replace(' Mei ','05')
df['Tanggal'] = df['Tanggal'].str.replace(' Juni ','06')
df['Tanggal'] = df['Tanggal'].str.replace(' Juli ','07')
df['Tanggal'] = df['Tanggal'].str.replace(' Agustus ','08')
df['Tanggal'] = df['Tanggal'].str.replace(' September ','09')
df['Tanggal'] = df['Tanggal'].str.replace(' Oktober ','10')
df['Tanggal'] = df['Tanggal'].str.replace(' November ','11')
df['Tanggal'] = df['Tanggal'].str.replace(' Desember ','12')
```

```python
df.head()
```

```python
#Merubah format kolom Tangal menjadi datetime
df['Tanggal'] = pd.to_datetime(df['Tanggal'], format='%d%m%Y')
```

```python
df.head()
```

```python
#Drop kolom yang tidak diperlukan
df.drop(['No.', 'Status', 'Pilihan'], axis = 1, inplace=True)
df.head()
```

```python
df.info()
```

```python
#Replace value yang kosong menjadi NaN
df = df.replace(r'^\s*$', np.NaN, regex=True)
```

```python
#Merubah tipe data kolom Nomor dan Jumlah Part menjadi numeric
df[['Nomor', 'Jumlah Part']] = df[['Nomor', 'Jumlah Part']].apply(pd.to_numeric)
```

```python
df.head()
```

```python
#Cek unique value kolom Jenis Layanan
df['Jenis Layanan'].unique()
```

```python
#Ubah value pada kolom Jenis Layanan dengan yang sesuai
df['Jenis Layanan'] = df['Jenis Layanan'].str.replace('PM pertama','PM 1 ')
df['Jenis Layanan'] = df['Jenis Layanan'].str.replace('3rd PM','PM 3 ')
```

```python
df['Jenis Layanan'].unique()
```

```python
#Cek unique value pada kolom Laporan Kerusakan
df['Laporan Kerusakan'].unique()
```

```python
#Ubah value pada kolom Laporan Kerusakan dengan yang sesuai
df['Laporan Kerusakan'] = df['Laporan Kerusakan'].str.replace('Preventive Maintenance ','PM')
df['Laporan Kerusakan'] = df['Laporan Kerusakan'].str.replace('preventif maintenes','PM')
```

```python
df['Laporan Kerusakan'].unique()
```

```python
df['CSE'].unique()
```

```python
#Simpan ke excel
df.to_excel('data-set-data aktifitas.xlsx')
```
